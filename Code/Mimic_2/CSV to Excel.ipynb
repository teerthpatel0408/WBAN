```python
import pandas as pd
import os
from glob import glob
```

```python
path_csv = r'MIMIC-II dataset\csv'
path_excel = r'MIMIC-II dataset\excel'
```

```python
# reading all the excel files 
csv_filenames = glob(path_csv + "\*.csv") 
for file in csv_filenames:
    print(file)
```

```python
for csv_file in csv_filenames:
    df = pd.read_csv(csv_file)
    excel_file = path_excel + '\\' + csv_file.split('\\')[-1].split('.')[0]+'.xlsx'
    print(excel_file)
    df.to_excel(excel_file, index=None, header=True)
```
